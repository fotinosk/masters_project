```python
import gym
import gym_Boeing
import numpy as np
import matplotlib.pyplot as plt
```

```python
env = gym.make('demonstration-v1')
```

```python
env.reset(ds=2)
```

```python
action = [0,0]

for i in range(2000):
    env.step(action)
```

```python
insp = env.inspect()
time = np.arange(0,len(insp)*0.05,0.05)
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams.update({'font.size': 18})

fig, ax = plt.subplots(nrows=3, ncols=1, sharex=True)
lbls = ['Vertical v.', 'Pitch angle', 'Pitch rate']
limits = [(-30,35), (-10,15), (0,85)]
for i in range(3):
    ax[i].plot(time,np.array(insp)[:,i+1])
    ax[i].set_ylabel(lbls[i])
    ax[i].set_ylim(limits[i])
ax[2].set_xlabel('Time (sec)')
plt.show()
```
